```python
def gugu(n):
    result = []
    i = 1
    while i< 10:
        result.append(n*i)
        i = i + 1
    return result
```

```python
print(gugu(2))
```

```python
def threeNfive(n):
    if n > 1000:
        print("안 되영")
    else:
        mylist = list(range(0, n+1))
        pluslist = []
        for i in mylist:
            if i % 3 == 0:
                pluslist.append(i)
            elif i % 5 == 0:
                pluslist.append(i)
            else:
                pass
        
        result = 0
        for j in pluslist:
            result += j
        
        print("0부터 입력한 숫자까지 3과 5의 배수 숫자들은")
        print(pluslist)
        print("이녀석들의 합은")
        print(result)
```
